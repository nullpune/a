#### 1. Import all the required Python Libraries. 

```python
import pandas as pd
import numpy as np

#to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#consider here Titanic-Dataset.csv
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)             # Fill missing with median
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  # Fill missing with mode
df.drop(columns=['Cabin'], inplace=True)                       # drop because more nullval
```

```python
df.head()
```

```python
df.dropna(subset=['Fare'], inplace=True) # droping row of missing value 
```

```python
df.drop(columns=['Ticket'], inplace=True) # droping column of missing value
```

```python
df['Survived'] = df['Survived'].astype('float')  # Ensure it's of integer type
print(f"\nAfter converting 'Survived' to int:\n{df['Survived'].dtype}")
```

```python
df['Sex'] = df['Sex'].astype('category')
print(f"\nAfter converting 'Sex' to category:\n{df['Sex'].dtype}")
```

```python
#after type conversion
df.dtypes
```

```python
#LABEL ENCODING

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
print(f"After Label Encoding, 'Sex' column:\n{df['Sex'].head()}")
```

```python
# One-Hot Encoding (for multi-category variables like 'Embarked' and 'Pclass')
# One-Hot Encoding for 'Embarked' (if it has multiple categories)

df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)  # drop_first=True to avoid multicollinearity
print(f"After One-Hot Encoding, the dataset:\n{df.head()}")

#df.head()
```
